```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.decomposition import PCA
```

```python
data = pd.read_csv("Titanic-Dataset.csv")
print(data.head())
print(data.info())
```

```python
data.fillna(data.mean(numeric_only=True), inplace=True)
data.fillna(data.mode().iloc[0], inplace=True)
```

```python
data.drop_duplicates(inplace=True)
```

```python
numeric_cols = data.select_dtypes(include=['number']).columns
Q1 = data[numeric_cols].quantile(0.25)
Q3 = data[numeric_cols].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outlier_condition = ((data[numeric_cols] < lower_bound) | (data[numeric_cols] > upper_bound)).any(axis=1)
data = data[~outlier_condition]
data
```

```python
data1 = pd.read_csv("Titanic-Dataset.csv")
data2 = pd.read_csv("Titanic-Dataset.csv")

df = pd.merge(data1, data2, on="PassengerId")
df
```
